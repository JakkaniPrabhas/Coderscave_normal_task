```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            return "Error: Division by zero is not allowed."
        else:
            return a / b

def display_menu():
    print("Select operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")

def get_user_input():
    while True:
        try:
            user_input = int(input("Enter your choice (1/2/3/4): "))
            if user_input >= 1 and user_input <= 4:
                return user_input
            else:
                print("Invalid choice. Please enter again.")
        except ValueError:
            print("Invalid choice. Please enter again.")

def get_operands():
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    return num1, num2

def main():
    calculator = Calculator()
    while True:
        display_menu()
        choice = get_user_input()

        if choice == 1:
            num1, num2 = get_operands()
            print("Result: ", calculator.add(num1, num2))
        elif choice == 2:
            num1, num2 = get_operands()
            print("Result: ", calculator.subtract(num1, num2))
        elif choice == 3:
            num1, num2 = get_operands()
            print("Result: ", calculator.multiply(num1, num2))
        elif choice == 4:
            num1, num2 = get_operands()
            print("Result: ", calculator.divide(num1, num2))
        else:
            print("Invalid choice.")

        next_calculation = input("Do you want to continue calculation (yes/no): ")
        if next_calculation == "no":
            break

if __name__ == "__main__":
    main()
```
